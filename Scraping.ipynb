### Data Scientists

```python
company_name=[]
cat_place = []
reviews = []
ratings = []
salaries = []
interviews = []
jobs = []
benefits = []

for i in range(1,8):
    
    url=f"https://www.ambitionbox.com/list-of-companies?designation=data-scientist&sortBy=topPaying&page={i}"
    page = requests.get(url,headers= req_headers)
    soup=BeautifulSoup(page.text,"html.parser")

    for card in soup.find_all("div", class_="companyCardWrapper"):

        primary = card.find("div", class_="companyCardWrapper__primaryInformation")
        secondary = card.find("div", class_="companyCardWrapper__tertiaryInformation")
    
        # ---- Primary div ----
        pname = primary.find("h2", class_="companyCardWrapper__companyName")
        company_name.append(pname.get_text(strip=True) if pname else np.nan)
    
        dplace = primary.find("span", class_="companyCardWrapper__interLinking")
        cat_place.append(dplace.get_text(strip=True) if dplace else np.nan)
    
        pl = primary.find("span", class_="companyCardWrapper__companyRatingCount")
        reviews.append(pl.get_text(strip=True) if pl else np.nan)
    
        rate = primary.find("div", class_="rating_text rating_text--md")
        ratings.append(rate.get_text(strip=True) if rate else np.nan)
    
        # ---- Secondary div -----
    
        counts = secondary.find_all("span", class_="companyCardWrapper__ActionCount")
        salaries.append(counts[1].get_text(strip=True) if len(counts) > 1 else np.nan)
        interviews.append(counts[2].get_text(strip=True) if len(counts) > 2 else np.nan)
        jobs.append(counts[3].get_text(strip=True) if len(counts) > 3 else np.nan)
        benefits.append(counts[4].get_text(strip=True) if len(counts) > 4 else np.nan)
```

```python
len(company_name)
```

```python
df_ds = pd.DataFrame({
    "company": company_name,
    "category_location": cat_place,
    "rating": ratings,
    "reviews": reviews,
    "salaries": salaries,
    "interviews": interviews,
    "jobs": jobs,
    "benefits": benefits
})

df_ds["designation"] = "Data Scientist"
```
